### Variables
We can store any value in a variable. A variable allows us to do two things:

Give a name to a value. This helps to improve the readability of our code.

Use the same value in multiple locations in our code without having to repeat it.

A variable can only ever contain one value at a time. To create a variable and assign it a value, write a variable name followed by the = sign and then the value to be stored there. A variable name can never be changed. However, the value inside of it can be changed by assigning a new value. In order to access a variable, write the name of the variable in the place where you want the value to be used. For example:


```python
my_favorite_color = 'blue'
print(my_favorite_color)
my_favorite_fruit = my_favorite_color + 'berries'
my_favorite_color = 'green'
print(my_favorite_color)
print(my_favorite_fruit)
```

### Strings
Strings are defined by a pair of either single or double quotes with text in between. Python has another special way of allowing longer, multi-line strings to be defined. You can use three double quotation marks """ at both the beginning and the end of a multi-line string. For example,


```python
long_string = """Hello this is a long string
and it goes on and on and on
across multiple lines
it ends here!"""

print(long_string)
```

To print a blank line between statements you can call the print() function with no input. For example:

```python
print('line1')
print()
print('line2')
```
